```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random

%matplotlib inline
```

# Data

```python
df = pd.read_csv('../data/RAW_recipes.csv')
```

```python
# since lists are not parsed correctly it must be evaluated
df["tags_list"] = df['tags'].apply(eval)
df["ingredients_list"] = df['ingredients'].apply(eval)

# this will let us search for the keywords we input the system
expl_tags = df['tags_list'].explode()
expl_ingredients = df['ingredients_list'].explode()

bathced_expl_lists = [expl_tags, expl_ingredients]
```

# Algorithm
- Based on tag and ingredients, filter for recipes that contain both

```python
# a simple algorithm, search for the filters in each corresponding category and then output the recipes
# at first it will be a list but later it will be a prompt, so it should recognize it
input_position_dict = {0: "tags", 1: "ingredients"}

tags_input = []
ingredients_input = ["soy sauce"]

batch_inputs = [tags_input, ingredients_input]

dict_indexes = {}


for e,inp in enumerate(batch_inputs):
    number_tag_inputs = len(inp)
    name_tag = input_position_dict[e]
    # print(name_tag)
    
    exploded_df = bathced_expl_lists[e]
    coinc_tags = exploded_df.loc[exploded_df.isin(inp)].index.value_counts()
    
    if len(coinc_tags) == 0:
        print(f"no coincidence for {input_position_dict[e]}")
    else:
        mask_presence_all = coinc_tags==number_tag_inputs
        if sum(mask_presence_all) == 0:
            # if there is but is lower
            
            # else then there aren't coincidences
            print("no coincidence")
        else:
            full_tag_indexes = coinc_tags[(coinc_tags==number_tag_inputs)].index
            dict_indexes[name_tag] = full_tag_indexes
            
chosen_recipes = list(set.intersection(*map(set,list(dict_indexes.values()))))
```
